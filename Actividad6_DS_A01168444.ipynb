3. Limpia los datos eliminando los registros nulos o rellena con la media de la columna

```python
# Creating a new dataframe
df1 = df.copy()

# We'll replace with mode for X2, X3, X4
# We'll replace with mean for X5
modeX2 = df1.X2.mode()
modeX3 = df1.X3.mode()
modeX4 = df1.X4.mode()
meanX5 = np.round(df1.X5.mean())

df1['X2'].fillna(value = modeX2, inplace = True)
df1['X3'].fillna(value = modeX3, inplace = True)
df1['X4'].fillna(value = modeX4, inplace = True)
df1['X5'].fillna(value = meanX5, inplace = True)

#Dropping any row with an empty value
df1.dropna(inplace = True)
df1HasNull = df1.isnull().values.any()
print("Empty values in DF1") if df1HasNull else print("No More NULLs in DF1")

#Lets compare sizes to check how many rows were dropped
print("Original Shape DF:",df.shape)
print("With Dropped Nulls DF1:",df1.shape)
```
