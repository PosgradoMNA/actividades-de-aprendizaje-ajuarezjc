# Notes on Linear Regression

## Simple Linear Regression

* One predictor and one target
* Based on a linear equation
* Set of training points, to fit the data into the model (train the model)
* Use model to predict y_hat (prediction)
* X and Y are usuarlly saved in array of numpy array

### Code for Simple Linear Regression

* Import the module
  * from sklearn.linear_model import LinearRegression
*Create object
  * lm = LinearRegression()
* Define predictor and target
  * X = df[predictor_column]
  * Y = df[target_column]
* Fit the data to the model, to find the parameters
  * lm.fit(X, Y)
* Obtain prediction
  * y_hat = lm.predict(X)
  * y_hat is an array with n as Xn inputs, but with the predicted values from the model

* We can get the coefficients of the model Y_hat = b0 + b1*X
  * Intercep (b0)
    * lm.intercept_
  * Slope (b1)
    * lm.coef_

## Multiple Linear Regression

* Two or more predictors, one continuos target
* Looking for equation given by: Y_hat = b0 + b1X1 + b2X2 + b3X3 ...

### Code for Multiple Linear Regression

* Import the module
  * from sklearn.linear_model import LinearRegression
*Create object
  * lm = LinearRegression()
* Define predictors and target
  * Z = df[[predictor_columns]]
  * Y = df[target_column]
* Fit the data to the model, to find the parameters
  * lm.fit(Z, Y)
* Obtain prediction
  * y_hat = lm.predict(X)
  * y_hat is an array with n as Xn inputs, but with the predicted values from the model

* We can get the coefficients of the model Y_hat = b0 + b1X1 + b2X2 + ...
  * Intercep (b0)
    * lm.intercept_
  * Slope (b1)
    * lm.coef_ : now an array. All coefficients from b1 to bn

## Model Evaluation using Visualization

  * Relationship between two variables
  * Strength of the correlation
  * Direction of the relationship (positive or negative)
* Regression plot shows a combination of 
  * Scatterplot: each point represents a Y point
  * The fitted linear regression line, represents the y_hat
* Horizontal axis is predictor, vertical axis is target

### Residual Plot

* We substract the value of Y_real from the Y_predict, and plotting the result according to the X axis. If the mean is 0, then linear regression was the right aproach.
* If we see the residual plot forms another pattern or the mean is not 0, then the Linear Regression might not be the right aproach
* If the variance of the residuals increases with X, then Linear Regression Model is incorrect too.

### Code for residual plot

* Import module
  * import seaborn as sns
* Create the plot
  * sns.residplot(df['preditor_col'], df['target_col'])

## Measures for InSample Evaluation

* Ways to numerically determine how good the predictions are

* Mean Squared Error (MSE)
* To obtain it with Python
  * from sklearn.metrics import mean_squared_error
  * mean_squared_error(y_real, y_hat)

* Rsquared / R2
  * Coefficient of Determination 
  * Measure to determine how close the data is to the fitted regression line
  * Percentage of variation of the target variable (Y) that is explained by the linear model
  * Usually between 0 and 1. Closer to 1 is better.
  * If R2 is negative, can be due to overfitting
  * To find using Python:
    * X = df['feature1']
    * Y = df['target']
    * lm.fit(X,Y)
    * lm.score(X, Y)
