## Notes on Module Evaluation

* In-sample evaluation tells how the model fits the data used to train it; it can't be used to tell how well the new data will be predicted
* We have to split our data up. One set to train and another as test data
* When splitting, the larger portion od data is used for training and the smaller part is used for testing. For example, 70/30 for training/testing

* We train the model with the training set.

* Using train_test_split function to automatically split datasets
  * Import the module
    * from sklearn.model_selection import train_test_split
  * The function returns the values: 
    * x_train, x_test, y_train, y_test
  * Use it
    * x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=0)
  * test_size defines the size of the test portion. The rest is used for train size

  * Remember to separate the data from the dataset.
    * x_data: features or independent variables
    * y_data: target

## Overfitting, underfitting and Model Selection section

* Underfitting: Model is under trained or too simple to fit the data
* Overfitting: Model fits the noise rather than the function
* Compare errors between models to find the model and features that better fit the model and predictions 

## Ridge Regression Section

* As polinomial order increases, we may need to use Ridge Regression to control the magnitudes of the higher order coefficients
* Ridge Regression introduces the parameter alpha to do so.
* Alpha is a parameter we select before training the model
* As alpha increases, the parameters of the coefficients get smaller. This is most evident for the higher order polynomial features
* If alpha is too large, the coefficients will tend to zero and the model will underfit the data
* If alpha is too little, the coefficients will tend to be big and the model will overfit the data
* We can use cross-validation to select an alpha value

* Using Ridge Regression

  * Import the module
    * from sklearn.linear_model import Ridge
  * Construct the model, here we specify the alpha value
    * RidgeModel = Ridge(alpha = 0.1)
  * Fit the data into the model
    * RidgeModel.fit(X, Y)
  * Obtain the predictios
    * y_hat = RidgeModel.predict(X)
  
# Grid Search Section

* Allows us to scan through multiple parameters, iterating the hyperparameters of different models
* To do this, we use the method Grid Search. It takes the model or objects to be trained and different values of the hyperparameters
* Then it calculates the mean square error or Rsquared for various hyperparameters
* Finally, GridSearch selects the hyperparameters that minimize the error

* The process uses a training set, validation set and test set
  * The model is trained with the training set, using different hyperparameters
  * Each model and combination of hyperparameters is validated on the validation set

* To know what parameters can be mixed and tested with GridSearch, we have to look at the documentation of each model

* In this module, we try with the polynomial model and the parameters alpha and normalization

* We create a list of parameters, where each element is a dictionary with key=name_of_parameter and value=[list_of_possible_values_for_the_parameter]
  * For example, to test multiple alphas we declare the list like:
    * parameters = [ {'alpha':[ 1, 10, 100, 1000 ]} ]
* R-squared is the default scoring method
    
* GridSeach takes in
  * List of parameters
  * Model
  * Scoring method
  * Number of folds for CrossValidation

* Gird search outputs:
  * List of different parameters and the resulting score
