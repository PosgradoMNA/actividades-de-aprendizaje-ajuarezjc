### Analysis of Variance
ANOVA (ANalysis Of VAriance)

To find the correlation between different groups of a categorical variable.
We obtain two values from ANOVA:
* F-test score: variation between sample group means divided by variation within sample group
* P-value: Confidence degree

* Small F implies poor correlation between variable categories and target variable

### Correlation

Statistical metric for measuring to what extent different variable are interdependant.

Positive correlation: As feature increases, output increases

Negative correlation: As feature increases, output decreases

Weak correlation: Feature changes but output remains similar

To obtain it:

sns.regplot(x='feature_col', y='output_column', data=df)
plt.ylim(0,)

### Correlation statistics

#### Pearson correlation
* Measure the strenght of the correlation between two feature
  * Correlation coefficient
  * P-value

* Correlation coefficient
  * Close to +1: Large Positive relationship
  * Close to -1: Large Negative relationship
  * CLose to 0: No relationship

* P-Value
  * < 0.0001 Strong certainty in the result
  * < 0.05   Moderate certainty
  * < 0.1    Weak certainty
  * >0.1     No certainty


* Example using  the Scipy package

pearson_coef, p_value = stats.pearsonr(df['feature'], df['output'])
